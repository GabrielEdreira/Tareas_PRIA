Realizar un programa en Python que pida al usuario la dimensión y los elementos de dos listas de números enteros (pueden tener distinta dimensión):

- Una vez leídos los datos y almacenados en las listas, se generará una nueva lista que será la intersección de ambas, es decir, contendrá aquellos elementos que se encuentren en ambas listas.
- El vector intersección se mostrará por pantalla siempre que existan elementos comunes.

NOTA: Se comprobará que el usuario no introduzca elementos repetidos dentro del mismo vector.


#### Ejemplo de ejecución:
<pre>
Introduce el numero de elementos de la lista 1 (>0):  3
Introduce los números:
Dato 1: 2
Dato 2: 4
Dato 3: 1

Introduce el numero de elementos de la lista 2 (>0):  5
Introduce los números:
Dato 1: 3
Dato 2: 3
Error, elemento repetido. Dato 2: 2
Dato 3: 9
Dato 4: 0
Dato 5: 1

La intersección de las listas es: 2 1

-----------------------------------------------------------

Introduce el numero de elementos de la lista 1 (>0):  2
Introduce los números:
Dato 0: 3
Dato 1: -2

Introduce el numero de elementos de la lista 2 (>0):  3
Introduce los números:
Dato 0: 0
Dato 1: 4
Dato 2: 5

No hay elementos comunes
</pre>

```python
# Realiza tu solución aquí o en un archivo .py independiente
def getList(name:str = "lista1") -> list:
    print(f"Introduce el numero de elementos de la {name} (>0):", end=" ")
    total_len = int(input(f"Introduce el numero de elementos de la {name} (>0):"))
    print(f"{total_len}\n")

    print("Introduce los números:")

    loop = True
    count = 0
    result = list()
    while(loop): 
        aux = -1
        while (aux < 0):
            print(f"Dato {count}:", end=" ")
            aux = int(input(f"Dato {count}").strip())
            print(f"{aux}")

            if (aux < 0):
                print("Valor incorrecto")

        if (result.__contains__(aux)):
            print(f"Error, elemento repetido. Dato: {count}")
        else:
            result.append(aux)
            count += 1

        if (len(result) == total_len):
            loop = False

    return result


def main():
    a = getList("lista1")
    b = getList("lista2")

    result = list()

    for a_item in a:
        if (result.__contains__(a_item)):
            continue

        for b_item in b:
            if (a_item == b_item):
                result.append(a_item)

    print(f"Los coincidentes son: {result}")

if __name__ == "__main__":
    main()


```

Escribir un programa que cree **un conjunto** simulando una lista de la compra. 

El programa debe preguntar el artículo y añadirlo al conjunto, mostrando un mensaje de aviso en caso de que el artículo ya existiese.



#### Ejemplo de ejecución:

<pre>
Introduce un artículo: pan
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: agua
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: leche
¿Quieres añadir artículos a la lista (Si/No)? Si
Introduce un artículo: café
¿Quieres añadir artículos a la lista (Si/No)? No
Artículos añadidos:
leche
agua
pan
café
Número de artículos:  4
</pre>

```python
# Realiza tu solución aquí o en un archivo .py independiente

def main():
    lista_compra = list()
    loop = True
    while (loop):
        print(f"Introduce un artículo:", end=" ")
        articulo = input(f"Introduce un artículo:").strip()
        print(f"{articulo}")
        
        print(f"¿Quieres añadir artículos a la lista (Si/No)?", end=" ")
        res = input(f"¿Quieres añadir artículos a la lista (Si/No)?").strip()
        print(f"{res}")

        if (res == "No"):
            loop = False
        else:
            lista_compra.append(articulo)

    print("Artículos añadidos:")
    for item in lista_compra:
        print(item)
    print(f"Número de artículos: {len(lista_compra)}")

if __name__ == "__main__":
    main()
```

```python
# Realiza tu solución aquí o en un archivo .py independiente

def main():
    print(f"Dime una frase:", end=" ")
    f = input(f"Dime una frase:").strip()
    print(f"{f}")

    c = None
    while (c == None or len(c) > 1):
        print(f"Dime un caracter:", end=" ")
        c = input(f"Dime un caracter:").strip()
        print(f"{c}")

    f = f.replace("a", c); f = f.replace("A", c.upper())
    f = f.replace("e", c); f = f.replace("E", c.upper())
    f = f.replace("i", c); f = f.replace("I", c.upper())
    f = f.replace("o", c); f = f.replace("O", c.upper())
    f = f.replace("u", c); f = f.replace("U", c.upper())

    print(f)

if __name__ == "__main__":
    main()



```

Realizar un programa en Python que permita gestionar las subscripciones de un usuario a diversas plataformas de entretenimiento, utilizando para ello **un diccionario**.

- El ejercicio parte de un diccionario que contiene el nombre y precio mensual de las subscripciones, formado por la clave (nombre de la plataforma) y el valor (precio mensual):


<pre>
        plataformas = {
        "Netflix": 11.99,
        "HBO": 8.99,       
        "Disney+": 7,
        "DAZN": 9.99,
        "Spotify": 9.99,
        "Apple Music": 9.99,
        "Amazon Prime": 3.99,
        }
</pre>


A continuación, deberá mostrar repetidamente un menú con las siguientes 6 opciones: 

1. Añadir plataforma (permite añadir una nueva plataforma al diccionario)
2. Eliminar plataforma (elimina una plataforma del diccionario)
3. Mostrar plataforma (muestra el precio de una plataforma a partir de su nombre)
4. Listar plataformas (muestra la lista de plataformas y sus precios)
5. Subscribirse n meses a una plataforma (muestra el precio total de la subscripción) 
6. Terminar (sale del programa)



#### Ejemplo de ejecución:

<img src="img/imagen1.png">

```python
# Realiza tu solución aquí o en un archivo .py independiente
plataformas = {
        "Netflix": 11.99,
        "HBO": 8.99,       
        "Disney+": 7,
        "DAZN": 9.99,
        "Spotify": 9.99,
        "Apple Music": 9.99,
        "Amazon Prime": 3.99,
        }

def printMenu():
    print(f"Menú de opciones\n (1) Añadir plataformas\n (2) Eliminar plataforma\n (3) Mostrar precio plataforma\n"
        + f" (4) Listar plataforma\n (5) Subscribirse n meses a una plataforma\n (6) Terminar\n Elige una opción:", end=" ")
    eleccion = int(input(f"Elección: ").strip())
    print(eleccion)

    return eleccion

def main():
    loop = True
    while (loop):
        match(printMenu()):
            case 1: # Añadir plataforma
                print("Indroduce el nombre de la nueva plataforma:", end=" ")
                nombre = input("Indroduce el nombre de la nueva plataforma: ")
                print(nombre)

                precio = 0
                while(precio <= 0):
                    print("Indroduce el precio de la plataforma:", end=" ")
                    try:
                        precio = int(input("Indroduce el precio de la plataforma: ").strip())
                    except:
                        precio = 0
                    print(precio)

                plataformas[nombre] = precio

            case 2: # Eliminar plataforma
                print()
                print("Indroduce el nombre de la plataforma:", end=" ")
                nombre = input("Indroduce el nombre de la plataforma: ")
                print(nombre)

                if (plataformas.__contains__(nombre)):
                    plataformas.pop(nombre)
                else:
                    print("La plataforma no existe")
            
                print()

            case 3: # Mostrar precio plataforma
                print()
                print("Indroduce el nombre de la plataforma:", end=" ")
                nombre = input("Indroduce el nombre de la plataforma: ")
                print(nombre)

                if (plataformas.__contains__(nombre)):
                    print(f"El precio mensual de {nombre} es de {plataformas[nombre]} €")
                else:
                    print("La plataforma no existe")

                print()

            case 4: # Listar plataforma
                print("\nPrecios mensuales de subscripción:")
                for key, item in plataformas.items():
                    print(f"{key}:{item}")
                print()

            case 5: #Subscribirse n meses
                print()
                print("Indroduce el nombre de la plataforma:", end=" ")
                nombre = input("Indroduce el nombre de la plataforma: ")
                print(nombre)
                
                num_meses = 0
                while(num_meses <= 0):
                    print("Indroduce el número de meses:", end=" ")
                    try:
                        num_meses = int(input("Indroduce el número de meses: ").strip())
                    except:
                        num_meses = 0
                    print(num_meses)
                
                print(f"El precio de subscripción de {num_meses} a {nombre} es de {plataformas[nombre] * num_meses} €")
                print()

            case 6: # Terminar
                loop = False
                print()
                print("¡Hasta pronto!")

            case _:
                print("Seleccione una opción válida")
if __name__ == "__main__":
    main()
    

```
