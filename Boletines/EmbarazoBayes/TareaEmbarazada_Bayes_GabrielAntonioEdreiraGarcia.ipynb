```python
import matplotlib.pyplot as plt
```

```python

def bayes_pregnancy(prior_p=0.15, sensitivity=0.9, false_positive=0.5, tests=5):
    probabilities = []
    p_e = prior_p  # Probabilidad inicial de estar embarazada
    
    for i in range(tests):
        p_t = (sensitivity * p_e) + (false_positive * (1 - p_e))
        p_e = (sensitivity * p_e) / p_t 
        probabilities.append(p_e * 100)
    
    return probabilities

num_tests = 5
results = bayes_pregnancy(tests=num_tests)

plt.figure(figsize=(8, 5))
plt.plot(range(1, num_tests + 1), results, marker='o', linestyle='-', color='g')
plt.xlabel("Positivos")
plt.ylabel("Probabilidad de estar embarazada")
plt.title("Proabilidad de embarazo positivo")
plt.grid()
plt.show()
```
