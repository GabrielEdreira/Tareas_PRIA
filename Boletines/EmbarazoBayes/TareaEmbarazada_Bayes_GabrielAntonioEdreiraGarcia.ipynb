```python
import matplotlib.pyplot as plt
```

```python

def Bayes(num, prior_p=0.15, sensitivity=0.9, falso_positivo=0.5):
    probabilidades = []
    p_e = prior_p
    
    for i in range(num):
        p_t = (sensitivity * p_e) + (falso_positivo * (1 - p_e))
        p_e = (sensitivity * p_e) / p_t 
        probabilidades.append(p_e * 100)
    
    return probabilidades

results = Bayes(4)

plt.figure(figsize=(8, 5))
plt.plot(range(1, 5), results, marker='o', linestyle='-', color='g')
plt.xlabel("Positivos")
plt.ylabel("Probabilidad de estar embarazada")
plt.title("Proabilidad de embarazo positivo")
plt.grid()
plt.show()
```
